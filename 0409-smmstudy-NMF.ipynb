# Binary NMF: MM & AL & POEMS
Ref: https://github.com/getzlab/SMM_clustering_2020
Cite: Bustoros, M., Anand, S., et al. "Genetic subtypes of smoldering multiple myeloma are associated with distinct pathogenic phenotypes and clinical outcomes." Nature communications 13.1 (2022): 1-10.

```python
import pandas as pd
import sys
import os
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
import seaborn as sns
import smm_utils as smm
import plot as pl
import signatureanalyzer as sa
```

```python
# Read input
X = pd.read_csv('./0429-snp-allmut-together-bi.csv',  index_col=0)
X=X.T
```

```python
H, W, _, _ = smm.bnmf(X.T, K=2, n_run=50, seed='nndsvd')
markers, signatures = sa.utils.select_markers(X.T, W, H, cut_norm=0.1, cut_diff=.1)

SUPP_DIR='./downsampling'
PLOTS_DIR='./figures'

H.to_csv(os.path.join(SUPP_DIR, "H_bnmf_k6_2.tsv"), sep='\t')
W.to_csv(os.path.join(SUPP_DIR, "W_bnmf_k6_2.tsv"), sep='\t')


_ = sa.pl.marker_heatmap(
    X.T,
    signatures,
    H['max_id'],
    diff=0,
    max_norm=0.1,
    display_y=True,
    figsize=(5,8),
    y_hm_label='',
    cmap='Greys'
)
plt.savefig(os.path.join(PLOTS_DIR, "bnmf_k6_2.pdf"), dpi=100, bbox_inches='tight')
```
